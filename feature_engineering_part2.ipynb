```python
home_data=pd.read_csv("X_train.csv")
```

```python
y_train=home_data[['SalePrice']]
```

```python
X_train=home_data.drop(['SalePrice','Id'],axis=1)
```

```python
X_train
```

Apply feature selection

1.first i specify the lasso regression model and 

2.select a suitable alpha

3.bigger the alpha the less feature that will be selected

4.then i selected model

5.will select the feature which coefficients are non-zero


```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
feature_sel_model=SelectFromModel(Lasso(alpha=0.005,random_state=0))
feature_sel_model.fit(X_train.fillna(0),y_train)
```

```python
feature_sel_model.get_support()
```

1.let is print the number of total and selected feature

2.this is how we can make a list of the selected feature




```python
import numpy as np
selected_feature=X_train.columns[(feature_sel_model.get_support())]
print("Total feature :{}".format((X_train.shape[1])))
print("Selected Features dataset :{}".format(len(selected_feature)))
```

```python
selected_feature
```

```python
X_train=X_train[selected_feature]
```
