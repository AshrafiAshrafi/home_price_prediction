let us capture all the nan values

first handle categorical feature which are missing

```python
categorical_nan=[feature for feature in home_data.columns if home_data[feature].isnull().sum()>1 and home_data[feature].dtype=='O']
```

```python
import numpy as np
for feature in categorical_nan:
    print("{} missing values {}% ".format(feature,np.round(home_data[feature].isnull().mean(),4)))
```

```python
def replacement(data,feature):
    data=home_data.copy()
    data[feature]=data[feature].fillna("Missing")
    return data
```

```python
home_data=replacement(home_data,categorical_nan)
```

```python
home_data[categorical_nan].isnull().sum()
```

```python
import numpy as np
for feature in categorical_nan:
    print("{} missing values {}% ".format(feature,np.round(home_data[feature].isnull().mean(),4)))
```

2.numerical nan missing values

```python
numerical_nan=[feature for feature in home_data.columns if home_data[feature].isnull().sum()>1 and home_data[feature].dtype!='O']
```

```python
for feature in numerical_nan:
    print("{} missing values{}%".format(feature,np.round(home_data[feature].isnull().mean(),4)))
```

```python
def numerical_missing(dataset,feature):
    dataset=dataset.copy()
    dataset[feature]=dataset[feature].median()
    return dataset    
```

```python
home_data=numerical_missing(home_data,numerical_nan)
```

```python
home_data[numerical_nan].isnull().sum()
```

```python
for feature in numerical_nan:
    print("{} missing values{}%".format(feature,np.round(home_data[feature].isnull().mean(),4)))
```

temporal variables (data time variables)

```python
for feature in['YearBuilt','YearRemodAdd','GarageYrBlt']:
    home_data[feature]=home_data['YrSold']-home_data[feature]
```

```python
categorical_feature=[feature for feature in home_data.columns if home_data[feature].dtype=='O']
```

```python
for feature in categorical_feature:
    temp=home_data.groupby(feature)['SalePrice'].count()/len(home_data)
    temp_df=temp[temp>0.01].index
    home_data[feature]=np.where(home_data[feature].isin(temp_df),home_data[feature],'Rare_var')
```

```python
home_data['MSZoning'].unique()
```

```python
for feature in categorical_feature:
    labels_ordered=home_data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    home_data[feature]=home_data[feature].map(labels_ordered)
```
